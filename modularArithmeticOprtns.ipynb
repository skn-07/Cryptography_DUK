```python
from findGCD import findGCD

def modularArithmetic(n1, n2, mod, opnd):
    if (opnd == 1):
        answer = (n1 + n2) % mod
    elif (opnd == 2):
        answer = (n1 * n2) % mod
    else:
        return False
    return answer

def modMultiplicativeInverse(num, modValue):
    gcd = findGCD(num, modValue)
    if (gcd == 1):
        i = 1
        while (True):
            modMult = modularArithmetic(num, i, modValue, 2)
            if (modMult == 1):
                return i
            else:
                i += 1
    else:
        return False

if __name__ == '__main__':
    number1 = int(input("Enter the first number:- "))
    number2 = int(input("Enter the second number:- "))
    modulus = int(input("Enter the modulus:- "))
    operand = int(input("Enter 1 for modular addition and 2 for modular multiplication:- "))
    ans = modularArithmetic(number1, number2, modulus, operand)
    if (ans == False):
        print("Invalid operand value!")
    elif (operand == 1):
        print(f"The modular addition of {number1} and {number2} by modulus {modulus} is {ans}.")
    elif (operand == 2):
        print(f"The modular multiplication of {number1} and {number2} by modulus {modulus} is {ans}.")

    num = int(input("Enter the number:- "))
    modValue = int(input("Enter the modulus:- "))
    modMultInverse = modMultiplicativeInverse(num, modValue)
    if (modMultInverse == False):
        print(f"{num} and {modValue} does not have any modular multiplicative inverse.")
    else:
        print(f"The modular multiplicative inverse of {num} and {modValue} is {modMultInverse}.")
```
