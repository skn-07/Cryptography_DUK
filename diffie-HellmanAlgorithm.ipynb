```python
from isPrime import isPrime
from primeFactors import primeFactors
import os

def findGenerator(number):
    if (number == 2):
        return 1
    else:
        eulersTotient = number - 1
        primeFactorsOfNum = list(primeFactors(eulersTotient))

        for i in range(2, number):
            isGenerator = 1
            for primes in primeFactorsOfNum:
                value1 = eulersTotient // primes
                value2 = pow(i, value1)
                finalValue = value2 % number
                if (finalValue == 1):
                    isGenerator = 0
                    break
            if (isGenerator == 1):
                generator = i
                print(f"Generator --> {generator}")
                return generator

if __name__ == "__main__":
    while (True):
        os.system('cls')
        print("Deffie-Hellman Algorithm")
        print("------------------------")
        primeNumber = int(input("Enter a prime number (0 to quit):- "))
        if (primeNumber == 0):
            break
        isNumPrime = isPrime(primeNumber)
        if (isNumPrime == 1):
            generatorOfNum = findGenerator(primeNumber)

            privateKeya = int(input("Enter the first private key for user 1 (number):- "))
            privateKeyb = int(input("Enter the second private key for user 2 (number):- "))

            publicKeyA = pow(generatorOfNum, (privateKeya % (primeNumber - 1)))
            publicKeyB = pow(generatorOfNum, (privateKeyb % (primeNumber - 1)))

            secretKey1 = pow(publicKeyB, (privateKeya % primeNumber))
            secretKey2 = pow(publicKeyA, (privateKeyb % primeNumber))

            if (secretKey1 == secretKey2):
                print("\nSecret keys match.")
            else:
                print("\nSecret keys does not match.")
            
            print(f"Secret key for user 1 is {secretKey1}.")
            print(f"Secret key for user 2 is {secretKey2}.")
            input("Press Enter to return to menu.")
        else:
            input("Number entered is not prime. Try again by pressing enter.")
```
