```python
from isPrime import isPrime

def primeFactors(number):
    hasFactors = isPrime(number)
    primeFactors = {}
    if (hasFactors == -1):
        return primeFactors
    elif (hasFactors == 1):
        primeFactors[1] = 1
        primeFactors[number] = 1
        return primeFactors
    else:
        num = number
        halfOfNum = number // 2
        while(num > 1):
            for i in range (2, halfOfNum + 1):
                if (num % i == 0):
                    if i in primeFactors:
                        primeFactors[i] += 1
                        break
                    else:
                        primeFactors[i] = 1
                        break
            num //= i
        return primeFactors

if __name__ == '__main__':
    number = int(input("Enter a number:- "))
    primeFact = primeFactors(number)
    if (len(primeFact) == 0):
        print("The number 1 does not have any prime factors.")
    else:
        print(f"The count of prime factors of {number} is {primeFact}.")
```
