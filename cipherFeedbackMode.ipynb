```python
from Cryptography.permutationCipher import sortBasedOnPermutation
from Cryptography.permutationCipher import inverseOfKey
import os

def cfbEncryption(plainText, key, IV, size):
    encryptedText = ""
    ivUsedForEnc = []
    keyLength = len(key)
    for i in range (0, len(plainText), keyLength):
        finalIV = IV[:size]
        block = plainText[i:i + keyLength]
        blockToString = "".join(str(i) for i in block)
        blockWithIV = int(blockToString, 2) ^ int(finalIV, 2)
        blockWithIVBinary = format(blockWithIV, 'b')
        blockWithIVBinaryPadded = blockWithIVBinary.zfill(keyLength)
        permutedBlock = sortBasedOnPermutation(blockWithIVBinaryPadded, key)
        encryptedBlock = "".join(permutedBlock)
        encryptedText += encryptedBlock
        IV = IV[size:] + encryptedBlock
        ivUsedForEnc.append(finalIV)
    
    return encryptedText, ivUsedForEnc

def cfbDecryption(cipherText, key, ivUsedForEnc):
    keyLength = len(key)
    decryptedText = ""
    ivIndex = 0
    inverseKey = inverseOfKey(key)
    for i in range (0, len(cipherText), keyLength):
        finalIV = ivUsedForEnc[ivIndex]
        block = cipherText[i:i + keyLength]
        blockToString = "".join(str(i) for i in block)
        permutedBlock = sortBasedOnPermutation(blockToString, inverseKey)
        permutedBlockToString = "".join(str(i) for i in permutedBlock)
        blockWithIV = int(permutedBlockToString, 2) ^ int(finalIV, 2)
        blockWithIVBinary = format(blockWithIV, 'b')
        blockWithIVBinaryPadded = blockWithIVBinary.zfill(keyLength)
        decryptedText += "".join(blockWithIVBinaryPadded)
        ivIndex += 1
    
    return decryptedText

if __name__ == '__main__':
    while(True):
        os.system('cls')
        print("Cipher Feedback Mode Cipher")
        print("---------------------------")
        print("1. Encrypt")
        print("2. Decrypt")
        print("3. Quit")
        option = int(input("Enter your choice (1-3):- "))
        if (option == 1):
            plainText = input("Enter the text to be encrypted:- ")
            key = input("Enter the key (Starting position is 1):- ")
            keyList = list(key)
            key = [int(i) for i in keyList]
            iv = input("Enter the iv:- ")
            size = int(input("Enter the size:- "))
            cipher, ivUsed = cfbEncryption(plainText, key, iv, size)
            print(f"The encrypted message is {cipher}.")
            print(f"The IV's used for encryption is {ivUsed}. **REQUIRED FOR DECRYPTION!**")
            input("Press Enter to continue...")
        elif (option == 2):
            cipher = input("Enter the text to be decrypted:- ")
            key = input("Enter the key (Starting position is 1):- ")
            keyList = list(key)
            key = [int(i) for i in keyList]
            ivUsed = []
            while(True):
                value = input("Enter the IV (Enter 'q' to Exit):- ")
                if (value.lower() == 'q'):
                    break
                else:
                    ivUsed.append(value)
            print(f"The encrypted message is {cfbDecryption(cipher, key, ivUsed)}.")
            input("Press Enter to continue...")
        elif (option == 3):
            break
        else:
            input("Wrong entry! Press Enter to retry...")
```
