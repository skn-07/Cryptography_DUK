```python
from permutationCipher import sortBasedOnPermutation
import os

def ctrEncryption(plainText, key, counter):
    keyLength = len(key)
    encryptedText = ""
    for i in range (0, len(plainText), keyLength):
        block = plainText[i:i + keyLength]
        blockToString = "".join(str(i) for i in block)
        blockWithCTR = int(blockToString, 2) ^ counter
        blockWithCTRBinary = format(blockWithCTR, 'b')
        blockWithCTRBinaryPadded = blockWithCTRBinary.zfill(keyLength)
        permutedBlock = sortBasedOnPermutation(blockWithCTRBinaryPadded, key)
        encryptedBlock = "".join(permutedBlock)
        encryptedText += encryptedBlock
        counter += 1
    return encryptedText

def ctrDecryption(cipherText, key, counter):
    keyLength = len(key)
    plainText = ""
    for i in range (0, len(cipherText), keyLength):
        block = cipherText[i:i + keyLength]
        blockToString = "".join(str(i) for i in block)
        blockWithCTR = int(blockToString, 2) ^ counter
        blockWithCTRBinary = format(blockWithCTR, 'b')
        blockWithCTRBinaryPadded = blockWithCTRBinary.zfill(4)
        decryptedBlock = sortBasedOnPermutation(blockWithCTRBinaryPadded, key)
        plainText += "".join(decryptedBlock)
        counter += 1
    return plainText

if __name__ == '__main__':
    while(True):
        os.system('cls')
        print("Counter Mode Cipher")
        print("-------------------")
        print("1. Encrypt")
        print("2. Decrypt")
        print("3. Quit")
        option = int(input("Enter your choice (1-3):- "))
        if (option == 1):
            plainText = input("Enter the text to be encrypted:- ")
            key = input("Enter the key (Starting position is 1):- ")
            keyList = list(key)
            key = [int(i) for i in keyList]
            counter = int(input("Enter the initial counter value:- "))
            print(f"The encrypted message is {ctrEncryption(plainText, key, counter)}.")
            input("Press Enter to continue...")
        elif (option == 2):
            cipher = input("Enter the text to be decrypted:- ")
            key = input("Enter the key (Starting position is 1):- ")
            keyList = list(key)
            key = [int(i) for i in keyList]
            counter = int(input("Enter the initial counter value:- "))
            print(f"The encrypted message is {ctrDecryption(cipher, key, counter)}.")
            input("Press Enter to continue...")
        elif (option == 3):
            break
        else:
            input("Wrong entry! Press Enter to retry...")
```
