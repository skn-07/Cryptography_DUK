```python
from rsaAlgorithm import charToBinary

def hashFunction1(string):
    binaryValue = charToBinary(string)
    fourBitSection = [binaryValue[start:start + 4] for start in range(0, len(binaryValue), 4)]
    hashedValue = int(fourBitSection[0], 2)
    for i in range(1, len(fourBitSection)):
        hashedValue = hashedValue ^ int(fourBitSection[i], 2)
    return format(hashedValue, '04b')

def hashFunction2(string):
    binaryValue = charToBinary(string)
    fourBitSection = [binaryValue[start:start + 4] for start in range(0, len(binaryValue), 4)]
    hashedValue = 0
    for i in range(0, len(fourBitSection)):
        hashedValue += i
        hashedValue %= 16
    return hashedValue

if __name__=="__main__":
    text = input("Enter a string:- ")
    hashValue1 = hashFunction1(text)
    print(f"1. The hashed value of {text} is {hashValue1}({chr(int(hashValue1, 2))}).")

    hashValue2 = hashFunction2(text)
    print(f"2. The hashed value of {text} is {hex(hashValue2)}({chr(hashValue2)}).")
```
